**Question 11.** Outlier-robust Huber regression

In a linear regression problem, we are given data $({\bf x}_i,y_i)\in \mathbb{R}^n \times \mathbb{R}$, $i=1,\ldots, m$. We fit an affine model

$$\hat{y}_i = {\bf x}_i^T{\boldsymbol \beta}$$

for $i=1,\ldots, m$ with ${\boldsymbol \beta} \in \mathbb{R}^n$.


In standard linear regression, we choose ${\boldsymbol \beta}$ by solving 
$$
\underset{{\boldsymbol \beta}}{\text{minimize}} 
\quad \sum\limits_{i=1}^m r_i^2 =  \sum\limits_{i=1}^m (y_i - {\bf x}_i^T{\boldsymbol \beta})^2. 
$$

Huber regression is a technique that is robust to outliers. In Huber regression, we replace the square function with the *Huber function*

$$
\phi(u) = \left\{ \begin{array}{ll}  u^2 & |u| \leq M\\
2M|u|  -  M^2 & |u| > M
\end{array}\right.
$$
where $M>0$ is the Huber threshold. That is, we solve

$$
\underset{{\boldsymbol \beta}}{\text{minimize}} \sum\limits_{i=1}^m \phi(y_i - {\bf x}_i^T{\boldsymbol \beta})
$$
for variable ${\boldsymbol \beta}.$

**[25 pts] Question 11c.**

We generate data with $m=500$ measurements and $n=300$ regressors.
We randomly choose ${\boldsymbol \beta}$ and ${\bf x}_i \sim \mathcal N(0,I)$.
We set $y_i = {\bf x}_i^T{\boldsymbol \beta} + \epsilon_i$, where $\epsilon_i \sim
\mathcal N(0,1)$. 
To introduce outliers, with probability $p$ we replace $y_i$ with $-y_i$. We generate $50$ problem instances, with $p$ varying from $0$ to $0.15$,


* Recover ${\boldsymbol \beta} \in {\bf R}^n$ from the measurements 
$y\in {\bf R}^m$ and compare the three approaches: linear regression, Huber regression and *clairvoyant* regression. In clairvoyant regression we know which measurements had their sign flipped.

* Plot the relative error 
$$
\frac{\|\hat{\boldsymbol \beta} - {\boldsymbol \beta}\|_2}{\|{\boldsymbol \beta}\|2},
$$
where $\hat{\boldsymbol \beta}$ is the recovered ${\boldsymbol \beta}$.

* Comment on your observations


Use [sklearn.linear_model](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.linear_model) for Huber regression and linear regression


```python
## Solution 11c.
import numpy as np
from sklearn.linear_model import HuberRegressor, LinearRegression
from numpy import linalg

## Generate data 
np.random.seed(1)
n = 300
m = 450
TESTS = 50

beta_true = 5*np.random.normal(size=(n,1))
X = np.random.randn(n, m)
Y = np.zeros((m,1))



p_vals = np.linspace(0,0.15, num=TESTS)
clairvoyant_error = np.zeros(TESTS)
huber_error = np.zeros(TESTS)
linear_error = np.zeros(TESTS)

# MODIFY and ENTER YOUR CODE HERE

for tst, p in enumerate(p_vals):
  for i in range(5):

    # Generate the sign changes.
    factor = 2*np.random.binomial(1, 1-p, size=(m,1)) - 1
    Y = factor*X.T.dot(beta_true)
    
    # print(tst)
    # print(p)


    # Solve standard regression 
    lr = LinearRegression()
    linear_regression= lr.fit(X.T, Y)
    beta_linear =linear_regression.coef_
    linear_error[tst]+=linalg.norm(beta_linear-beta_true.T,2)/linalg.norm(beta_true.T,2)

      # Solve clairvoyant regression 

    y_clair=np.multiply(factor,Y)
    #print("Y_Clav",y_clair.shape)
    clairvoyant_reg=LinearRegression().fit(X.T,y_clair)
    beta_clairvoyant=clairvoyant_reg.coef_
    clairvoyant_error[tst]+=linalg.norm(beta_clairvoyant-beta_true.T,2)/linalg.norm(beta_true.T,2)

    # Solve Huber regression  with M=1
    huber_reg=HuberRegressor(epsilon=1,max_iter=10000,alpha=0.0).fit(X.T,Y.flatten())
    beta_huber=huber_reg.coef_
    beta_huber=beta_huber.reshape(n,1)
    huber_error[tst]+=linalg.norm(beta_huber-beta_true.T,2)/linalg.norm(beta_true.T,2)
  huber_error[tst]/=5   
  clairvoyant_error[tst]/=5
  linear_error[tst]/=5


     

```

```python
# Plot the relative error for the three methods
# for p_vals
from matplotlib import pyplot as plt
from matplotlib import axes 


# plt.semilogy(years, indexValues )

# plot.ylim([10,21000])

# plot.xlim([1900,2020])
fig ,axis=plt.subplots(4,figsize=(10,10))

axis[0].semilogy(p_vals,linear_error,label="linear",color="blue")
axis[0].semilogy(p_vals,clairvoyant_error,label="clair",color="red")
axis[0].semilogy(p_vals,huber_error,label="huber",color="orange")




axis[1].plot(p_vals,linear_error,label="linear",color="blue")
axis[2].plot(p_vals,clairvoyant_error,label="clair",color="red")
axis[3].plot(p_vals,huber_error,label="huber",color="orange")
```
